```python
# Importar bibliotecas necesarias
import pandas as pd
import altair as alt

# Cargar el archivo CSV
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Base-de-datos-.csv')

data.head(2)
```

```python
#Limpiar y procesar los datos
#Eliminar columnas que no aportan valor a la visualización
# Fix: Corrected column names to match the actual DataFrame columns (case-sensitive)
columnas_relevantes = ['Cantantes femeninas en Chile','Canta musica urbana', 'Edad','Ano de exito']
datos_filtrados = data[columnas_relevantes].dropna()

#Crear un gráfico de barras básico para visualizar cantantes femeninas en chile por ano de exito
chart = alt.Chart(datos_filtrados).mark_bar().encode(
    # Fix: Corrected column name in the x encoding
    x=alt.X('Cantantes femeninas en Chile',sort='-y', title= 'cantantes femeninas en chile'),
    y=alt.Y('Ano de exito',title ='Ano de exito'),
    tooltip=['Cantantes femeninas en Chile','Ano de exito']
).properties(
    title='cantantes femeninas por ano de exito',
    width=600,
    height=400
)
chart

#Guardar el grafico como archivo HTML para visualizarlo en un navegador
chart.save('visualización1.html')

#mostrar un mensaje indicando que el archivo se ha creado
print("el grafico se ha guardado como 'visualizacion1.html'. Abrélo en navegador para visualizacion.")


```

```python
# Importar bibliotecas necesarias
import pandas as pd
import altair as alt

# Cargar el archivo CSV
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Base-de-datos-.csv')

# Limpiar y procesar los datos
# Eliminar columnas que no aportan valor a la visualización
columnas_relevantes = ['Cantantes femeninas en Chile', 'Edad', 'Canta musica urbana']
datos_filtrados = data[columnas_relevantes].dropna()

# Crear selección interactiva
selection = alt.selection_interval()

# Gráfico de puntos para visualizar Compliance Score por Control
points = alt.Chart(datos_filtrados).mark_point().encode(
    x='Cantantes femeninas en Chile:N',
    y='Edad:Q',
    tooltip=['Cantantes femeninas en Chile', 'Edad', 'Canta musica urbana'],
    color=alt.condition(selection, 'Cantantes femeninas en Chile:N', alt.value('lightgray'))
).add_selection(
    selection
).properties(
    width=600,
    height=400,
    title='Edades de las cantantes femeninas en Chile'
)
# Histograma para visualizar la cantidad de controles por conjunto
bars = alt.Chart(datos_filtrados).mark_bar().encode(
    y='Cantantes femeninas en Chile:N',
    color='Cantantes femeninas en Chile:N',
    x='count Edad:Q'
).transform_filter(
    selection
    ).properties(
    width=600,
    height=400,
    title='Cantidad de edades de las cantantes chilenas'
)

# Combinar los gráficos
final_chart = points & bars

# prueba seleccionando un cuadrado en el gráfico de puntos para que el gráfico de barras cambie sus valores

final_chart

# Guardar el gráfico combinado como archivo HTML
final_chart.save('visualizacion2.html')
# Mostrar un mensaje indicando que el archivo ha sido creado
print("El gráfico se ha guardado como 'visualizacion2.html'. Ábrelo en el navegador para guardarlo")

```
